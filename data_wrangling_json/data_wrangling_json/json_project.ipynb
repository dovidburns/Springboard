## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## Question 1: Find the 10 countries with most projects. 
# Answer 

```python
# We start by reading in the json data as a dataframe and looking at the columns in it to find the column with the country names.
json_df = pd.read_json('data/world_bank_projects.json')
json_df.columns
```

```python
# We see above the country names column is named "countryname". Since each row in this dataset is a different project,
# we can use a groupby function to count the number of rows there are for each country name. We then sort these values in
# a descending way and only take the top 10 results to find the 10 countries with the most projects. 
df =json_df 
df =df.groupby(df.countryname).size()
df.sort_values(ascending=False).head(10)
```

```python
# An alternative to using groupby is simply using the built in pandas function value_counts on the column countryname to get 
# the number of projects each country has. Then taking the top 10 using a .head(10) call will give the desired outcome.
df = json_df.countryname.value_counts()
df.head(10)
```

## Question 2: Find the top 10 major project themes (using column 'mjtheme_namecode')
# Answer 

```python
# We first create a pandas dataframe and look at what is in the column named 'mjtheme_namecode'.
df = pd.read_json('data/world_bank_projects.json')
df.mjtheme_namecode[1]
```

```python
# We see above that in the column 'mjtheme_namecode', we have a nested list with both a 'code' and a coresponding number
# and a 'name' with the corresponding name. We can then use the json_normalize function on this data to separate these out 
# into columns named 'code' and 'name' with the corresponding project theme code and names. 
data = json.load((open('data/world_bank_projects.json')))
df = json_normalize(data, 'mjtheme_namecode')
df
```

## Questions 3: In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.
# Answer 

```python
# The theory here is to create a kind of data frame key based on correlating the code to the name and then applying that back 
# to the data frame for all the codes to fill in the missing names. 
```

```python
# We start by sorting all the values by code. 
data = json.load((open('data/world_bank_projects.json')))
df = json_normalize(data, 'mjtheme_namecode')  
df.code = df.code.astype('int') # changes the code column to intiger format so that sort_values works properly on it.   
df = df.sort_values("code")
df
```

```python
# We then need to get a full list of the project names in order of code. 
data = json.load((open('data/world_bank_projects.json')))
df = json_normalize(data, 'mjtheme_namecode') 
df.code = df.code.astype('int') # changes the code column to intiger format so that sort_values works properly on it. 
df = df.sort_values("code")

names = list(df['name'].unique()) #  grabs each unique name and the blank name, only once and puts them in a list. 
del names[0] # removes the blank name
names
```

```python
# Using the same method above we get an ordered list of the product codes as well. 
data = json.load((open('data/world_bank_projects.json')))
df = json_normalize(data, 'mjtheme_namecode')  
df.code = df.code.astype('int')
df = df.sort_values("code")

codes = list(df.code.unique())


codes
```

```python
# Using the above lists of codes and names we create a dataframe with just each unique code and name. We then set the index 
# to be the code number. This will allow us to create a lookup function for a project name based on its corresponding code. 
d ={'codes':codes,'names':names}
table = pd.DataFrame(data=d)
table2 = table.set_index(['codes'])
table2
```

```python
# This cell just verifies that by for an X in code numbers, calling table2["names"][X] will give that codes corresponding name.
table2["names"][3]
```

```python
# a reminder of the original data frame with missing names
df.head(20)
```

```python
# Verify the length of our data frame with missing names to compare to new data frome to be created below. 
len(df)
```

```python
# We can now run through our dataframe code column and search table2 for the correct name to apply based on the codes in our 
# dataframe. This is then saved to a new dataframe which does not have any missing values in the names column. 
df_filled_in = df.drop('name',1)
df_filled_in["name_filled"] = df["code"].apply(lambda x : table2["names"][x] )
df_filled_in
```

```python
# Verifying that we now have more counts in each column as the missing names have been filled in. 
df_filled_in.name_filled.value_counts()
```
